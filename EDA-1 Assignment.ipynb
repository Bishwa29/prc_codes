```python
##Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

##A6. We need to follow the below steps:

##Preprocess the Data: Scale the features for PCA.
##Perform PCA: Fit PCA on the scaled data.
##Analyze Explained Variance: Determine how many components are needed to reach 90% variance explained.
```

```python
##Preprocess the data
##Scaling the features is essential for PCA, as it is sensitive to the scale of the data.

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('winequality-red.csv')

# Separate features and target
X = df.drop(columns=['quality'])  # Features
y = df['quality']  # Target (not used in PCA)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python

# Perform PCA
pca = PCA()
pca.fit(X_scaled)

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Cumulative explained variance
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Print cumulative explained variance for each component
for i, var in enumerate(cumulative_explained_variance):
    print(f'Component {i+1}: Cumulative Explained Variance = {var:.4f}')
```

```python
# Find the number of components that explain at least 90% of the variance
num_components = np.argmax(cumulative_explained_variance >= 0.90) + 1

print(f'Minimum number of components to explain at least 90% of the variance: {num_components}')
```

```python
##Plot the cumulative explained variance to visualize the results.

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.axhline(y=0.90, color='r', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance vs. Number of Principal Components')
plt.grid(True)
plt.show()
```
